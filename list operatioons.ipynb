```python
#1  creating a list
skills=["python","sql","ai"]
numbers=[10,20,30,40,50]
print(skills)
print(numbers)
#2 accesing elements
#positive index
print(skills[0])
#negative index
print(skills[-2])
```

```python
#3 slicing lists
print(numbers[1:4])
print(numbers[:3])
print(numbers[4:])
print(numbers[1:6:2])
print(numbers[0:1:3])
print(numbers[::2])
```

```python
#4 inserting elements
skills=["python","sql","ai"]
print(skills.insert(1,'git'))
print(skills)

#5 modifying elements
skills[1]="power bi"
print(skills)

#6 adding elements
skills.append("fast api")
print(skills)
```

```python
#7 extending list
a=[1,2]
b=[3,4]
a.extend(b)
print(a)

#8 concat list
a=[1,2]
b=[3,4]
c=a+b
print(c)
```

```python
#9 removing  elements
skills=["python","sql","ai"]
skills.remove("sql")
print(skills)

#10 removing last element
skills=["python","sql","ai"]
skills.pop()
print(skills)

#11 pop using index
skills.pop(1)
print(skills)

#12 delete using del
del skills[0]
print(skills)
```

```python
#13 clear center list
skills=["python","sql","ai"]
skills.clear()
print(skills)

#14 find length
print(len(numbers))

#15 check existence
print("python" in skills)
numbers=[10,20,30,40,50]
print(10 in numbers)

#16 count occurance
numbers=[1,1,2,3]
print(numbers.count(1))

#17 find index
numbers=[10,20,30,40,50]
print(numbers.index(20))
```

```python
#18 sort list
values=[50,20,80,97,54]
values.sort()
print(values)
values.sort(reverse=True)
print(values)
```

```python
#19 reverse list
values=[50,20,80,97,54]
values.reverse()
print(values)
values[::-1]

#20 copy list
new_list=values.copy()
print(new_list)
```

```python
#21 loop through list 
values=[50,20,80,97,54]
for value in values:
    print(value)

#22 loop through index
for i  in range(len(values)):
    print(i,values[i])
```

```python
#23 list comprehension
numbers=[1,2,3]
squares=[x*x for x in numbers]
print(squares)

#24 nested lists
students=[['rahul','sai'],['harsha','shaistha']]
print(students[0][0])
print(students[0][1])
print(students[1][0])
print(students[1][1])

#25 loop through nested list
for s in students:
    print(s)

#26 nested loop 
for  student in students:
    for  value in students:
        print(value)
```

```python
#27 multiple list operations
names=["rahul","janu"]
roles=["ai","backend"]
for n,r in zip(names,roles):
    print(n,r)

#28 find max/min
numbers=[5,2,7]
print(max(numbers))
print(min(numbers))

#29 sum  of list
print(sum(numbers))
```

```python
#30 split string into list
skills="python sql ai"
skills_list=skills.split()
print(skills_list)

#31 join list into string
skills=["python",'sql']
print(" , ".join(skills))

#32 join
skills='python in backend lang'
age='22'
print(skills.join('ai'))
```

```python
#string to list operations
college="my college is good"
college_bio=college.split()
print(college_bio)

# list to string
college_bio=['my','college','is','good']
college=" ".join(college_bio)
print(college)
```
